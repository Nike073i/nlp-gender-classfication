```python
import pandas as pd
import os

docs_dir = './data/rusid'
output_dir = './data'
output_file = os.path.join(output_dir, 'source.xlsx')

xlsx_files = [file for file in os.listdir(docs_dir) if file.endswith('.xlsx') and file.startswith('blogs')]

cols = ['Текст: Возраст автора', 'Текст: Текст', 'Автор: Пол', 'Корпус: Название']
combined_df = pd.DataFrame()

for file in xlsx_files:
    file_path = os.path.join(docs_dir, file)
    df = pd.read_excel(file_path)
    data = df[(df['Текст: Форма речи'] == 'монолог') & (df['Автор: Родной язык'] == 'русский')]
    combined_df = pd.concat([combined_df, data[cols]], ignore_index=True)

combined_df.drop_duplicates(inplace=True)
combined_df.dropna(inplace=True)
combined_df.columns = ['Возраст', 'Текст', 'Пол', 'Корпус']
combined_df.to_excel(output_file)
```
